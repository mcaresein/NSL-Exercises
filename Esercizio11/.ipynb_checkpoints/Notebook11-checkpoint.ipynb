### Exercise 11.1

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects


m = 2 # slope
b = 1 # intersect
for sigma in [0, 0.2]:
    print('==================================== sigma = '+str(sigma)+' ====================================')
    for Ntrain in [100, 150, 200]:
        i=0
        plt.figure(figsize=(16,8))
        for epchs in [30, 50, 70]:
            i=i+1
            np.random.seed(0)
            x_train = np.random.uniform(-1, 1, Ntrain)
            x_valid = np.random.uniform(-1, 1, 10)
            x_valid.sort()
            y_target = m * x_valid + b # ideal (target) linear function
            print(x_train.shape)
            #sigma = 0.0 # noise standard deviation, for the moment it is absent
            y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
            y_valid = np.random.normal(m * x_valid + b, sigma)

            # plot training and target dataset
            #plt.plot(x_valid, y_target)
            #plt.scatter(x_valid, y_valid, color='r')
            #plt.grid(True); plt.show()

            model = tf.keras.Sequential()
            model.add(Dense(1, input_shape=(1,)))

            model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

            model.summary()

            history = model.fit(x=x_train, y=y_train, 
                                batch_size=32, epochs=epchs,
                                shuffle=True, # a good idea is to shuffle input before at each epoch
                                validation_data=(x_valid, y_valid),
                                verbose=0 )

            #model.get_weights()

            score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)
            print()
            print('Test loss:', score[0])
            print('Test accuracy:', score[1])

            # evaluate model with the exact curve
            score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
            print()
            print('Test loss:', score[0])
            print('Test accuracy:', score[1])
            
            # Plot training & validation loss values
            plt.subplot(2,3,i)
            plt.plot(history.history['loss'])
            plt.plot(history.history['val_loss'])
            plt.title('Model loss')
            plt.ylabel('Loss')
            plt.xlabel('Epoch')
            plt.legend(['Train', 'Test'], loc='best')

            plt.subplot(2,3,3+i)
            x_predicted = np.random.uniform(-1, 1, 100)
            y_predicted = model.predict(x_predicted)
            plt.scatter(x_predicted, y_predicted,color='r')
            plt.plot(x_valid, y_target)
            plt.grid(True)
            plt.tight_layout
        plt.show()
```

From the above plots comes out that both increasing the training set and the epoch results in better prediction performances of the neural network. It's worth noting that increasing the number of epoch without providing a sufficiently large dataset could cause overfitting. 
Moreover, clearly in this particularly trivial case adding noise to the data degrades the performances

### Exercise 11.2

```python

def new_model(neurons_per_layer, layers, activation_fn):
    
    model = Sequential()
    model.add(Dense(neurons_per_layer, input_shape=(1,), activation=activation_fn))
       
    for i in range(layers):
        model.add(Dense(neurons_per_layer, activation=activation_fn,))
                    
                    
    model.add(Dense(1))

    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
    return model


a = 3
b = 2
c = 3
d = 4

#generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 5000)
x_valid = np.random.uniform(-1, 1, 500)
x_valid.sort()
y_target = a * x_valid**3 - b * x_valid**2 - c * x_valid + d


#studia come cambia premettendo i segni giusti o sbagliati

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal( a *  x_train**3 - b * x_train**2 - c * x_train + d ) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal( a *  x_valid**3 - b * x_valid**2 - c * x_valid + d )

#plot training and target dataset
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show();
plt.clf()

Nneurons_per_layer=[3, 5, 7, 10, 20]
nlayers=[1, 2, 3]

for layers in nlayers:
    i=0
    plt.figure(figsize=(18, 9))

    for neurons_per_layer in Nneurons_per_layer:
        i=i+1
        model=new_model(neurons_per_layer, layers, "relu")

        #get a summary of our composed model
        #model.summary()

        history = model.fit(x=x_train, y=y_train,batch_size=32, epochs=50,
                    shuffle=True,validation_data=(x_valid, y_valid), verbose=0)

        model.get_weights()

        #evaluate model
        score_valid = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

        #print performance
        #print()
        #print('Test loss:', score[0])
        #print('Test accuracy:', score[1])

        #evaluate model with the exact curve
        score_target = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)
        
        #print performance
        print()
        #print('Test loss:', score[0])
        #print('Test accuracy:', score[1])

        #A trained model can be saved and stored to a file for later retreival. This allows you to checkpoint a model and resume training later without rebuilding and training from scratch.
        #Files are saved in HDF5 format, within all weight values, model's configuration and even the optimizer's configuration.# look into training history
        #Plot training & validation loss values

        plt.subplot(3,len(Nneurons_per_layer),i)
        plt.plot(history.history['loss'])
        plt.plot(history.history['val_loss'])
        plt.title( str(neurons_per_layer)+' Neurons per '+str(layers)+' layer\n'
                  'v_loss = '+str(score_valid[0])+'\n'+'t_loss = '+str(score_target[0]) )
        plt.ylabel('Loss')
        plt.xlabel('Epoch')
        plt.legend(['Train', 'Test'], loc='best')

        plt.subplot(3,len(Nneurons_per_layer),5+i)
        x_predicted = np.random.uniform(-1, 1, 50)
        y_predicted = model.predict(x_predicted)
        plt.scatter(x_predicted, y_predicted,color='r')
        plt.plot(x_valid, y_target)
        plt.title("Prediction on the same training domain")
        plt.grid(True)
        
        plt.subplot(3,len(Nneurons_per_layer),10+i)
        x_predicted = np.random.uniform(-1.5, 1.5, 50)
        y_predicted = model.predict(x_predicted)
        plt.scatter(x_predicted, y_predicted,color='r')
        xext_valid = np.random.uniform(-1.5, 1.5, 50)
        xext_valid.sort()
        yext_target = a * xext_valid**3 - b * xext_valid**2 - c * xext_valid + d
        plt.plot(xext_valid, yext_target )        
        plt.title("Prediction on enlarged domain")
        plt.grid(True)

        plt.tight_layout()
    plt.show()
```

### Exercise 11.3
  
Try to extend the model to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects


np.random.seed(0)
Ntrain = 5000
Nvalid = 500
sigma  = 0.0 
epochs = 50

x_train = np.random.uniform(-1.5, 1.5, Ntrain)
y_train = np.random.uniform(-1.5, 1.5, Ntrain)
#data_train = np.meshgrid(x_train, y_train).
data_train = np.array([x_train , y_train]).T

z_train = np.random.normal( np.sin(x_train**2 + y_train**2),sigma )
print(data_train.shape)
print(np.array(z_train).shape)


x_valid = np.random.uniform(-1.5, 1.5, 100) 
x_valid.sort()
y_valid = np.random.uniform(-1.5, 1.5, 100) 
y_valid.sort()
data_valid = np.array([x_valid, y_valid]).T
z_valid  = np.random.normal( np.sin(x_valid**2 + x_valid**2) )
z_target = np.sin(x_valid**2 + x_valid**2)

# plot training and target dataset
#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
#plt.grid(True);
#ax.plot_wireframe(data_valid[0], data_valid[1], z_target)

model = tf.keras.Sequential()
model.add(Dense(15, input_shape=(2,), activation="relu") )
model.add(Dense(15, activation="relu"))
#model.add(Dense(35, activation="relu"))
#model.add(Dense(35, activation="relu"))
model.add(Dense(1))
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()

earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_loss',
                  min_delta=0, patience=100, mode='auto')

history = model.fit(x=data_train, y=z_train, 
                    batch_size=32, epochs=epochs,
                    shuffle=True, # a good idea is to shuffle input before at each epoch
                    validation_data=(data_valid, y_valid),
                    verbose=1,callbacks=earlystop )

#model.get_weights()
#
score = model.evaluate(data_valid, z_valid, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(data_valid, z_target, batch_size=32, verbose=1)
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# Plot training & validation loss values
#plt.figure(figsize=(16,8))

#plt.subplot(1,2,1)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')

#plt.subplot(1,2,2)
fig = plt.figure(figsize=(19,9))

ax = fig.add_subplot(121, projection='3d')
ax.view_init(0, 10)
x_predicted = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = np.random.uniform(-1.5, 1.5, 1000)
data_predicted=np.array([x_predicted, y_predicted]).T
z_predicted = model.predict(data_predicted)
ax.scatter3D(data_predicted[:,0],data_predicted[:,1], z_predicted,color='r', s=0.5)
X, Y=np.meshgrid(x_valid,y_valid);
ax.plot_wireframe(X, Y, np.sin(X**2 + Y**2),linewidth=2, alpha=0.1)

ax = fig.add_subplot(122, projection='3d')
ax.view_init(80, 0)
ax.scatter3D(data_predicted[:,0],data_predicted[:,1], z_predicted,color='r', s=0.5)
ax.plot_wireframe(X, Y, np.sin(X**2 + Y**2),linewidth=2, alpha=0.1)
plt.grid(True);

plt.show()
```
