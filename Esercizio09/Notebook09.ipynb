The simulations are carried on with a population of $150$ individual, and the paths shown below are the best path in the $150th$ complete new generations.

## Cities on a square

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(16,8))
x, f = np.loadtxt("square/bestpath_square.txt", unpack=True)
plt.plot(x,f, label="Best Path")

x, f = np.loadtxt("square/squareworld", unpack=True, skiprows=1)
plt.scatter(x,f, color='brown', label="Cities")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(16,8))
x, y = np.loadtxt("square/output_bestdistance.dat", unpack=True, usecols=[0,1])
plt.plot(x,y,label='$R^1_{best path}$')

x, y ,err = np.loadtxt("square/rstatistics", unpack=True, usecols=[0,1,2])
plt.errorbar(x,y,yerr=err,label=(r'$\langle R^1\rangle_{\{best\,\,half\}}$') )

plt.xlabel('blocks')
plt.ylabel('$R^1$ Distance')

plt.show()
```

The best path produced by the simulations is clearly not the absolute best, because just rearranging the path in order to avoid the two crossings can reduce the travelling distance. As shown in the Fig. 2, the candidate optimal path saturate for high blocks, and this is why I have to generate much more new generation in order to increase to find a best candidate.
